```python
import pandas as pd
```

#### I got dataset from the website of Istanbul Municipality, available to public.
https://data.ibb.gov.tr/dataset/sehir-tiyatrolari-veri-seti

```python
data = pd.read_csv("theater_play.csv", sep=",")
data
```

#### I will discover distinct names of theaters and their longitudes and latitudes

```python
data.groupby(["THEATER_NAME"]).THEATER_NAME.unique()
```

```python
arr1 = data.groupby(["THEATER_NAME"]).LONGITUDE.unique()
arr1
```

```python
df1 = pd.DataFrame(arr1)
df1
```

```python
arr2 = data.groupby(["THEATER_NAME"]).LATITUDE.unique()
arr2
```

```python
df2 = pd.DataFrame(arr2)
df2
```

```python
all_data = df1.merge(df2, how="inner",on="THEATER_NAME")
all_data
```

#### I am cleaning closed brackets from location values. Theater names are indices of the table.

```python
all_data['LONGITUDE'] = all_data['LONGITUDE'].str.get(0)
all_data['LATITUDE'] = all_data['LATITUDE'].str.get(0)
all_data
```

```python
all_data.reset_index(inplace=True)
all_data
```

#### Before saving to csv file, I will delete indents at begining of theater names.

```python
all_data['THEATER_NAME'] = all_data['THEATER_NAME'].str.strip()
all_data
```

```python
all_data.to_csv("tiyatro.csv",index=False,sep=";")
```

```python
fdir = r"C:\Users\berka\OneDrive\Belgeler\10real_world_applications\basemap\tiyatro.csv"
```

```python
data = pd.read_csv(fdir,sep=";")
data
```

```python
sahne = list(data["THEATER_NAME"])
lat = list(data["LATITUDE"])
lon = list(data["LONGITUDE"])
```

```python
fg = folium.FeatureGroup(name="My Map")
```

```python
for shn,lt,ln in zip(sahne,lat,lon):
    fg.add_child(folium.Marker(location=[lt,ln],
                                popup=shn, 
                                icon=folium.Icon(color="blue")))
```

```python
map.add_child(fg)
```

```python
map.save("city_theaters.html")
```
